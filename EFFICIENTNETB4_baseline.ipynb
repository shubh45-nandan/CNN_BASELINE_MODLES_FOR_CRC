```python

ZIP_IN_DRIVE = "/content/drive/MyDrive/archive (2).zip"

# ---- Installs ----
!pip install -q tensorflow keras opencv-python matplotlib seaborn scikit-learn

# ---- Mount Drive ----
from google.colab import drive
drive.mount('/content/drive')

# ---- Imports ----
import os, glob, shutil, random, json, sys
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from collections import Counter
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications import EfficientNetB4
from tensorflow.keras.applications.efficientnet import preprocess_input
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, GlobalAveragePooling2D, BatchNormalization
from tensorflow.keras import regularizers

# ---- Optimizer (no addons needed) ----
AdamW_available = True
try:
    from keras.optimizers import AdamW  # Keras 3+
except:
    AdamW_available = False
    from tensorflow.keras.optimizers import Adam

print("âœ… Libraries ready")

# ==========================================================
# ğŸ“¦ ALWAYS USE ZIP DATASET
# ==========================================================
if not ZIP_IN_DRIVE or not os.path.exists(ZIP_IN_DRIVE):
    raise FileNotFoundError("âŒ ZIP_IN_DRIVE path is not set correctly. Fix it at the top.")

EXTRACT_DIR = "/content/Extracted_Kather"
if os.path.exists(EXTRACT_DIR):
    shutil.rmtree(EXTRACT_DIR)
os.makedirs(EXTRACT_DIR, exist_ok=True)

print(f"ğŸ“¦ Unzipping ZIP dataset:\nâ†’ {ZIP_IN_DRIVE}")
!unzip -q -o "$ZIP_IN_DRIVE" -d "$EXTRACT_DIR"

# Auto-detect dataset root (the folder containing class subfolders)
def find_dataset_root(base):
    # Try deepest folder containing most subfolders
    best = base
    best_count = 0
    for root, dirs, files in os.walk(base):
        if len(dirs) > best_count:
            best = root
            best_count = len(dirs)
    return best

SOURCE_DIR = find_dataset_root(EXTRACT_DIR)
print(f"âœ… Detected dataset root: {SOURCE_DIR}")

# ==========================================================
# âš™ï¸ Reproducibility + Hyperparams
# ==========================================================
SEED = 42
random.seed(SEED); np.random.seed(SEED); tf.random.set_seed(SEED)

IMG_SIZE = (380, 380)
BATCH_SIZE = 32
EPOCHS_INITIAL = 15
EPOCHS_FINE = 25

VAL_RATIO = 0.20
TEST_RATIO = 0.15

EXTS = (".jpg", ".jpeg", ".png", ".bmp", ".tif", ".tiff",
        ".JPG", ".JPEG", ".PNG", ".BMP", ".TIF", ".TIFF")

print("âœ… Basic config loaded.")

# ==========================================================
# ğŸ” Discover classes with images
# ==========================================================
def list_images_recursive(folder):
    out = []
    for root, _, files in os.walk(folder):
        for fn in files:
            if os.path.splitext(fn)[1] in EXTS:
                out.append(os.path.join(root, fn))
    return sorted(out)

class_to_files = {}
for cls in sorted(os.listdir(SOURCE_DIR)):
    p = os.path.join(SOURCE_DIR, cls)
    if os.path.isdir(p):
        files = list_images_recursive(p)
        if files:
            class_to_files[cls] = files

available_classes = sorted(class_to_files.keys())
missing_classes = []

if len(available_classes) == 0:
    raise RuntimeError("âŒ No images found. Wrong ZIP structure?")

print("\nğŸ“š Classes detected:")
for c in available_classes:
    print(f"  - {c}: {len(class_to_files[c])} images")

# ==========================================================
# ğŸ§© Build Train/Val/Test Stratified Split
# ==========================================================
SPLIT_ROOT = "/content/kather_split"
TRAIN_DIR = SPLIT_ROOT + "/train"
VAL_DIR   = SPLIT_ROOT + "/val"
TEST_DIR  = SPLIT_ROOT + "/test"

if os.path.exists(SPLIT_ROOT):
    shutil.rmtree(SPLIT_ROOT)

os.makedirs(TRAIN_DIR); os.makedirs(VAL_DIR); os.makedirs(TEST_DIR)

def allocate_counts(n, val_ratio=VAL_RATIO, test_ratio=TEST_RATIO):
    if n <= 0: return (0,0,0)
    if n == 1: return (1,0,0)
    if n == 2: return (1,0,1)

    test_n = max(1, int(n * test_ratio))
    val_n  = max(1, int(n * val_ratio))
    if test_n + val_n >= n:
        val_n = 1; test_n = 1
    train_n = n - val_n - test_n
    return (train_n, val_n, test_n)

totals = {"train":0,"val":0,"test":0}
rng = np.random.default_rng(SEED)

for cls in available_classes:
    files = class_to_files[cls]
    idx = np.arange(len(files))
    rng.shuffle(idx)

    train_n,val_n,test_n = allocate_counts(len(files))
    train_idx = set(idx[:train_n])
    val_idx   = set(idx[train_n:train_n+val_n])
    test_idx  = set(idx[train_n+val_n:train_n+val_n+test_n])

    os.makedirs(f"{TRAIN_DIR}/{cls}")
    os.makedirs(f"{VAL_DIR}/{cls}")
    os.makedirs(f"{TEST_DIR}/{cls}")

    for i in idx:
        src = files[i]
        base = os.path.basename(src)
        if i in train_idx: dst = f"{TRAIN_DIR}/{cls}/{base}"
        elif i in val_idx: dst = f"{VAL_DIR}/{cls}/{base}"
        else:              dst = f"{TEST_DIR}/{cls}/{base}"
        shutil.copy2(src, dst)

    totals["train"] += train_n
    totals["val"]   += val_n
    totals["test"]  += test_n

print("\nâœ… Split completed:")
print(totals)

# ==========================================================
# ğŸ”„ Generators
# ==========================================================
train_datagen = ImageDataGenerator(
    preprocessing_function=preprocess_input,
    rotation_range=25,
    width_shift_range=0.2, height_shift_range=0.2,
    zoom_range=0.3,
    horizontal_flip=True, vertical_flip=True,
    brightness_range=[0.8,1.2],
    shear_range=0.2,
    fill_mode="reflect"
)

val_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)
test_datagen = ImageDataGenerator(preprocessing_function=preprocess_input)

train_gen = train_datagen.flow_from_directory(TRAIN_DIR, target_size=IMG_SIZE,batch_size=BATCH_SIZE,shuffle=True,class_mode='categorical')
val_gen   = val_datagen.flow_from_directory(VAL_DIR,   target_size=IMG_SIZE,batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical')
test_gen  = test_datagen.flow_from_directory(TEST_DIR, target_size=IMG_SIZE,batch_size=BATCH_SIZE,shuffle=False,class_mode='categorical')

class_names = [None]*len(train_gen.class_indices)
for name,idx in train_gen.class_indices.items():
    class_names[idx] = name
num_classes = len(class_names)

# ==========================================================
# ğŸ§  Build EfficientNetB4
# ==========================================================
base_model = EfficientNetB4(weights='imagenet',include_top=False,input_shape=(*IMG_SIZE,3))
base_model.trainable = False

l2_reg = regularizers.l2(1e-4)

model = Sequential([
    base_model,
    GlobalAveragePooling2D(),
    BatchNormalization(),
    Dense(512, activation='relu', kernel_regularizer=l2_reg),
    Dropout(0.5),
    BatchNormalization(),
    Dense(256, activation='relu', kernel_regularizer=l2_reg),
    Dropout(0.3),
    Dense(num_classes, activation='softmax')
])

opt1 = AdamW(learning_rate=1e-3, weight_decay=1e-4) if AdamW_available else Adam(1e-3)
model.compile(optimizer=opt1,loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.1),metrics=['accuracy'])

print(model.summary())

callbacks1 = [
    tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=5,restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',factor=0.5,patience=3),
    tf.keras.callbacks.ModelCheckpoint('/content/best_phase1.h5',save_best_only=True)
]

print("\nğŸ¯ Phase 1 training...")
hist1 = model.fit(train_gen,validation_data=val_gen,epochs=15,callbacks=callbacks1)

# ---------------------------
# ğŸ”“ Phase 2 Fine-tuning
# ---------------------------
base_model.trainable = True
for layer in base_model.layers[:150]:
    layer.trainable = False

opt2 = AdamW(learning_rate=1e-5, weight_decay=1e-4) if AdamW_available else Adam(1e-5)
model.compile(optimizer=opt2,loss=tf.keras.losses.CategoricalCrossentropy(label_smoothing=0.05),metrics=['accuracy'])

callbacks2 = [
    tf.keras.callbacks.EarlyStopping(monitor='val_accuracy',patience=8,restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss',factor=0.5,patience=4),
    tf.keras.callbacks.ModelCheckpoint('/content/best_phase2.h5',save_best_only=True)
]

print("\nğŸ”“ Phase 2 training...")
hist2 = model.fit(train_gen,validation_data=val_gen,epochs=25,callbacks=callbacks2)

# ==========================================================
# ğŸ§¾ Evaluation Function
# ==========================================================
def evaluate(gen,name):
    print(f"\nğŸ§ª Evaluating on {name}...")
    model.load_weights('/content/best_phase2.h5')

    loss,acc = model.evaluate(gen,verbose=0)
    print(f"âœ… {name} Accuracy = {acc*100:.2f}%")

    y_true = gen.classes
    y_pred = np.argmax(model.predict(gen,verbose=0),axis=1)

    print(classification_report(y_true,y_pred,target_names=class_names,digits=4))
    cm = confusion_matrix(y_true,y_pred)
    plt.figure(figsize=(8,7))
    sns.heatmap(cm,annot=True,fmt='d',cmap='Blues',xticklabels=class_names,yticklabels=class_names)
    plt.title(f"{name} Confusion Matrix")
    plt.show()

evaluate(val_gen,"Validation")
evaluate(test_gen,"Test")

print("\nğŸ‰ ALL DONE â€” ZIP DATASET TRAINING COMPLETE âœ…")
```
